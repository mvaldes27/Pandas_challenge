```python
# Dependencies and Setup
import pandas as pd

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

## Player Count

```python
total_players = len(purchase_data_df["SN"].unique())

total_players_dict = {"Total Players": [total_players] }
                       
total_players_df = pd.DataFrame(total_players_dict)

total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data_df["Item Name"].unique())
avg_price = round(purchase_data_df["Price"].mean(), 2)
number_of_purchases = len(purchase_data_df["Purchase ID"].unique())
total_revenue = purchase_data_df["Price"].sum()

purchase_summary = {
                "Number of Unique Items" : [unique_items],
                "Average Price" : [avg_price],
                "Number of Purchases" : [number_of_purchases],
                "Total Revenue" : [total_revenue]
}

purchase_summary_df = pd.DataFrame(purchase_summary)

purchase_summary_df['Total Revenue'] = f'${total_revenue}'
purchase_summary_df['Average Price'] = f'${avg_price}'

purchase_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Filter out duplicates
clean_gender_df = purchase_data_df.drop_duplicates(subset=['SN', 'Gender'])


#Find totals by category

total_people = len(clean_gender_df)
total_male = len(clean_gender_df.loc[purchase_data_df["Gender"] == "Male", :])
total_female = len(clean_gender_df.loc[purchase_data_df["Gender"] == "Female", :])
total_other = len(clean_gender_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :])


#Find percentages by category

percent_male = total_male / total_people
percent_female = total_female / total_people
percent_other = total_other / total_people

#summarize findings 
gender_summary = {
                "Total Count" : [total_male,total_female, total_other ], 
                "Percentage of Players" : [percent_male, percent_female, percent_other]
                
        
}

gender_summary_df = pd.DataFrame(gender_summary)

gender_summary_df_new = gender_summary_df.rename(index={0 : 'Male', 1 : 'Female' , 2 : 'Other / Non-Disclosed'})

gender_summary_df_new["Percentage of Players"] = gender_summary_df_new["Percentage of Players"].map("{:.2%}".format)

gender_summary_df_new

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#find purchases by each person
purchases_by_SN =purchase_data_df["SN"].value_counts()

#find total purchase value for each SN
sn_df = purchase_data_df.groupby(["SN"]).sum().sort_values("Price", ascending=False)
del sn_df['Age']
del sn_df['Purchase ID']
del sn_df['Item ID']
#rename column Price to Total Purchase Value
renamed_sn_df = sn_df.rename(columns={"Price":"Total Purchase Value"})

#add purchase count to dataframe
renamed_sn_df['Purchase Count'] = purchases_by_SN.values

#convert datatypes of columns to numeric
renamed_sn_df['Purchase Count'] = pd.to_numeric(renamed_sn_df['Purchase Count'])
renamed_sn_df['Total Purchase Value'] =  pd.to_numeric(renamed_sn_df['Total Purchase Value'])

#find average purchase price by diving total purchase value by purchase count
Avg_PP = renamed_sn_df['Total Purchase Value'] / renamed_sn_df['Purchase Count']

#add Avg Purchase Price to dataframe
renamed_sn_df['Average Purchase Price'] = Avg_PP.values

#reorganize columns
organized_sn_df = renamed_sn_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

#format numbers
organized_sn_df["Average Purchase Price"] = organized_sn_df["Average Purchase Price"].map("${:.2f}".format)
organized_sn_df["Total Purchase Value"] = organized_sn_df["Total Purchase Value"].map("${:.2f}".format)
organized_sn_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create new df with items information
items = purchase_data_df[["Item ID", "Item Name", "Price"]]
items
#group item data by item id and item name
item_info = items.groupby(["Item ID", "Item Name"])
#add purchase count
purchase_count_item = item_info["Price"].count()
#count purchase value per item
purchase_value_item = (item_info["Price"].sum())
#find individual item price
item_price = purchase_value_item/purchase_count_item

#most popular items
most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item,
                                   "Item Price": item_price, 
                                   "Total Purchase Value": purchase_value_item})
#sort in descending order
most_popular_sorted = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()

#format numbers
most_popular_sorted["Item Price"] = most_popular_sorted["Item Price"].map("${:.2f}".format)
most_popular_sorted["Total Purchase Value"] = most_popular_sorted["Total Purchase Value"].map("${:.2f}".format)
most_popular_sorted
```
